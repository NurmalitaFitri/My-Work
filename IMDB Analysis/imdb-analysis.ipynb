```python
import pandas as pd
```

```python
df = pd.read_csv('../input/imdb-extensive-dataset/IMDb movies.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

## Data Cleansing

```python
df
```

```python
df.drop_duplicates(inplace = True, keep = 'last')
df.head()
```

## EDA

```python
# 10 TOP Movies with high average vote
df[['original_title','year','genre','avg_vote','production_company']].sort_values(by=['avg_vote'], ascending=False).head(10)
```

```python
#10 Movies with highest reviews from users 
df[['original_title','year','genre','avg_vote','reviews_from_users','country','production_company',]].sort_values(by=['reviews_from_users'], ascending=False).head(10)
```

```python
#5 Country with the most production movie
df[['country','original_title']].groupby(['country'])['original_title'].count().nlargest(5)
```

```python
#Rating with the duration length
df[['original_title','year','duration','avg_vote']].sort_values(by=['avg_vote'], ascending = False).head(10)
```

```python
#Movie with the largest budgets
df[['original_title','year','genre','budget','country','production_company']].sort_values(by=['budget'], ascending = False).head()
```

```python
#5 movie with the highest worldwide gross income
df[['original_title','year','worlwide_gross_income','budget','production_company']].sort_values(by=['worlwide_gross_income'], ascending = False).head()
```

```python
gb_df = df.groupby(['original_title', 'genre','year']).agg({'country' : 'sum', 'avg_vote' : 'sum'})
gb_df.head(10)
```

## Data Visualization

```python
#The number of 5 Country with the most production movie
df[['country','original_title']].groupby(['country'])['original_title'].count().nlargest(5).plot.barh(figsize=(18,5))
```

```python
#10 highest rating movie
df[['production_company','avg_vote']].groupby(['production_company'])['avg_vote'].sum().nlargest(10).plot.barh(figsize=(18,5))
```

```python
#10 writers with the highest average vote for the movies
df[['writer','avg_vote']].groupby(['writer'])['avg_vote'].sum().nlargest(10).plot.bar(figsize=(18,5))
```
